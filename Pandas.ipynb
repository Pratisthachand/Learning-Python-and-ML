```python
#Creating a series from a list:

import pandas as pd

a= [0,3,5,7,9]
s1= pd.Series(a) #first column gives order and the other column gives data
print (s1)

order2=[1,2,3,4,5] # so order starts from 1 instead of 0 now
s2= pd.Series(a, index=order2) #Series must be capital 
print (s2)

order3=['a','b','c','d','e'] #order can be letters too
s3= pd.Series(a,index=order3)
print (s3)
```

```python
import pandas as pd

a= [2,4,6,8]
order= ('A','B','C','D')

s1=pd.Series(a,index=order)
print (s1)

# Another way to modify the index of series:

a.index= ['p','q','r','s'] #no of objects in index must be = no of objects in data
s2=pd.Series(a)
print s2

```

```python
import pandas as pd

a= {'a':1,'b':2,'c':3}  #assigning index to data 
s1=pd.Series(a)
print (s1)
```

```python
#Slicing:

import pandas as pd

a=[1,2,3,4,5]
s1=a[:3] # prints from start to the third object (1st three)
print (s1)

s2=a[2:] #prints objects after 2 objects (excludes 2nd object)
print (s2)

s3=a[:-1] #prints everything except for last object
print (s3)

s4=a[-2:] # prints everything except first two
print (s4)
```

```python
#Joining two series:
import pandas as pd

a= {'a':1,'b':2,'c':3,'d':4}
s1=pd.Series(a)
b= {'p':5,'q':6,'r':7,'s':8}
s2=pd.Series(b)

s3= s1.append(s2)
print (s3)

s3.drop('q') #deletes index q and its data. doesn't change original b tho.
```

```python
#Series operations

import pandas as pd

arr1=[0,1,2,3,4,5]
arr2=[6,7,8,9]

s1= pd.Series(arr1)
s2= pd.Series(arr2)


#+-*/ series:

s3= s1.div(s2) #can replace div with sum or sub or mul 
print (s3)

print('median', s1.median())
print('max', s1.max())
print('min', s1.min())
```

```python
#Creating Dataframe
#A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

import pandas as pd
import numpy as np

dates=pd.date_range('today',periods=6) #Define time sequence as index
num_arr= np.random.randn(6,4) #import numpy random array
columns=['A','B','C','D']#Use the table as the column name

df1=pd.DataFrame(num_arr, index=dates, columns=columns)

print (df1)

```

```python
#Create dataframe with dictionary array

import pandas as pd
data= {'animal':['cat','cat','snake','dog','dog','cat','snake','cat','dog','dog'],
       'age':[2.5,3,0.5,np.nan, 5,2,4.5,np.nan,7,3], #np.nan= no value --we can include this to make the no of objects in all lists equal.
       'visits':[1,3,2,3,2,3,1,1,2,1],
       'priority':['yes','yes','no','yes','no','no','no','yes','no','no']}
labels=['a','b','c','d','e','f','g','h','i','j']

df2=pd.DataFrame(data, index=labels)
print(df2)

print (" ")

print (df2.dtypes) 

print (" ")

df3=df2.head(6) #prints top 6 values
print (df3)

print (" ")

df4=df2.tail(3) #prints last 3 values
print (df4)

print (" ")
print(df2.columns)

print (" ")
print (df2.values)

print (" ")
print(df2.describe()) #see statistical data of dataframe

print (" ")
print (df2.T) #swaps the column and rows 

print (" ")
df2.sort_values(by='age') #sorts data by age in increasing order'

print (" ")
print (df2[1:3]) #slicing dataframe by row (included 2 and 3 data)

print (" ")

#query dataframe by tag
print (df2[['age','visits']]) #shows only data under age and visits column 

print (" ")

#identical to slicing= integer location (iloc)

print (df2.iloc[1:3]) # prints rows 2,3 only
print (" ")

#
df3=df2.copy()
print (df3.isnull()) #creates chart of null values #true = having null value

print (" ")

#modyfying location:

df3.loc['f','age']=1.5 #changed f ko age from 2 to 1.5
print (df3)

print (" ")

#
print(df3.mean())

print (" ")

print(df3.sum())
```

```python
#OPERATIONS FOR DATAFRAME MISSING VALUES:

#filling missing values:
import pandas as pd
data= {'animal':['cat','cat','snake','dog','dog','cat','snake','cat','dog','dog'],
       'age':[2.5,3,0.5,np.nan, 5,2,4.5,np.nan,7,3], #np.nan= no value --we can include this to make the no of objects in all lists equal.
       'visits':[1,3,2,3,2,3,1,1,2,1],
       'priority':['yes','yes','no','yes','no','no','no','yes','no','no']}
labels=['a','b','c','d','e','f','g','h','i','j']

df1=pd.DataFrame(data, index=labels)

df2=df1.copy()
print (df2)

df2.fillna(11) #fills NaN values with 11 in df2 



```

```python
#VISUALIZATION IN PANDAS

import pandas as pd
import numpy as np
import matplotlib as mpt

ts=pd.Series(np.random.randn(50), index= pd.date_range('today', periods=50))
ts=ts.cumsum() #aka cumulative sum
ts.plot()

df=pd.DataFrame(np.random.randn(50,4),index=ts.index, columns=['A','B','P','Q'])
df=df.cumsum()
df.plot()
```

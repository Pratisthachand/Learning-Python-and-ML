```python
#Import the functions from Keras and skLearn libraries

import math
from keras.models import Sequential #Keras provides a number of model builing APIs like Sequential, Functional and Subclassing. Sequential - going 1 tensorflow layer at a time, one direction at a time(ex: left to right or top to bottom). Functional- complicated directions. SUbclassing- very complicated.

#Importing layers from keras: (dense and LSTM)
#Layers = basic building blocks of neural networks in Keras

from keras.layers import Dense #Dense layer is a standard neural network. 
from keras.layers import LSTM #Long Short Term Memory Layer

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
df= pd.read_csv('/content/air_quality.csv',sep=' ')   #no need to import pandas again. just make sure to run the code cell where we have imported pandas. 
df.head()
```

```python
#CONVERT TIME TO DATETIME FORMAT:

#this will combine the two columns of time(Tanggal) and date(Jam).

def combine_date(df, tab_name):
  list_tab = []
  for i in range(df.shape[0]):
    list_tab.append(df.loc[i, 'Tanggal']+ 'T' + df.loc[i, tab_name][0:2])
  return np.array(list_tab, dtype='datetime64') 
df['Datetime']= combine_date(df,'Jam')

#Convert into hourly data:
df = df[['Datetime','O3','CO', 'NO2','NO','CO2', 'VOC', 'PM1', 'PM2.5', 'PM4',	'PM10',	'TSP',	'TEMP',	'HUM',	'WS',	'WD',	'ISPU']]

df2= df.groupby(['Datetime']).mean()
df2.head()

#we see a new column called "Datetime" where both time and date is present.
```

```python
#Description statistics of the variables:

df2.describe()
```

```python
#Calculate the quantile for each variable:

def calculate_quantile (i,df2):
  Q1 = df2[[i]].quantile(0.25)[0]
  Q3 = df2[[i]].quantile(0.75)[0]
  IQR = Q3- Q1
  min = df2[[i]].min()[0]
  max = df2[[i]].max()[0]
  min_IQR = Q1 - 1.5*IQR
  max_IQR = Q3 + 1.5*IQR

  return Q1, Q3, min, max, min_IQR, max_IQR

#delete first and last rows to avoid missing value extrapolation:

df2.drop(index=[df2.index[0], df2.index[df2.shape[0]-1]], inplace= True)

#find and interpolate the outliers:

for i in df2.columns:
  print('\nAttribute-',i,':')
  Q1, Q3, min, max, min_IQR, max_IQR = calculate_quantile(i, df2)
  print('Q1 = %.2f' % Q1) #%.2f means to round up to two decimal places
  print('Q3  =%.2f' % Q3)
  print('min IQR = %.2f' % min_IQR)
  print('max IQR = %.2f' % max_IQR)
  if (min< min_IQR):
    print('---> Low outlier is found = %.2f'% min)
  if (max> max_IQR):
    print('---> High outlier is found =%.2f'% max)

#CLEARING OUTLIERS OUT OF THE DATA: 

  def convert_nan (x, max_IQR=max_IQR, min_IQR=min_IQR):
    if ((x > max_IQR) | (x < min_IQR)): # | stands for and
      x = np.nan    #if condition is true, x is given null value     
    else:
      x = x         #else x will remain the same
    return x

  def convert_nan_HUM (x, max_IQR= 100, min_IQR=min_IQR):
    if ((x > max_IQR) | (x < min_IQR)): 
      x = np.nan      
    else:
      x = x         
    return x

  if (i == 'HUM'): #HUM =humidity, aka has weird values. so we're clearing it out
    df2[i] = df2[i].map(convert_nan_HUM)
    df2[i] = df2[i].interpolate(method='linear') #interpolated means we've given them a new value
  if (i != 'HUM'): # != means not equal to
    df2[i]= df2[i].map(convert_nan)
    df2[i]= df2[i].interpolate(method='linear') #method= linear means that we are filling the data in based on the linear regression of similar data

  if (len(df2[df2[i].isnull()][i]) == 0):
    print ('### Outliers have been interpolated!!! ###')


#INDENTATION MAKES A LOT OF DIFFERENCE ON THE OUTCOME OF THE CODE!!!!
```

```python
#Log transformation to deal wiht skewed data:

dataset= np.log1p(df2[['TEMP']].values)
dataset.shape
```

```python
#Create new dataframe to compare the original vs Log transform data:

dist_df= pd.DataFrame ({'TEMP': df2['TEMP'].values, 'log_TEMP': dataset[:,0]})  #dist_df means distribution dataframe

#HISTOGRAM plot original vs LOG Transform data:
plt.figure(figsize=(12,5))
dist_df.hist();
```

```python
#SPLITTING the dataset into train and test sets:

train_size = int(len(dataset)* 0.75) #len means length. so train size is 75% of dataset. The len function returns the length of the string, which happens to be an integer.
test_size = len(dataset)- train_size
train, test = dataset[0:train_size, :], dataset[train_size:len(dataset),:]
print (len(train), len(test))
```

```python
#Convert an array of values into a Dataset matrix:

def create_dataset (dataset, look_back = 1): #look back 1 piece of data going backward.
  dataX, dataY = [],[]
  for i in range(len(dataset)- look_back-1):
      a = dataset [i:(i+ look_back),0]
      dataX.append(a)
      dataY.append(dataset[i + look_back,0])
  return np.array(dataX), np.array(dataY)

#Reshape the train and test data:

look_back =1
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)

print ('Shape of trainX :', trainX.shape)
print ('Shape of trainY :', trainY.shape)
print ('Shape of testX :', testX.shape)
print ('Shape of testY :', testY.shape) #As trainY has only one column, this causes some problems so we have to reshape the input array.

#RESHAPE the input array in the form - [samples, time steps, features]

      #(NOTES: LSTM take into account the past data in addition to the the current data point in order to make “contextual” and more accurate predictions. Timesteps means how much behind do you want to look into to make the prediction OR how many past data points do you want to consider in addition to the current data point to make the prediction.For example, if the timestep is set to 50, it means that the previous 50 datapoints and the current datapoint will be used to make the prediction.

trainX= np.reshape (trainX, (trainX.shape[0],1,trainX.shape[1]))
testX = np.reshape(testX, (testX.shape[0],1 ,testX.shape[1]))

print ('Shape of trainX :', trainX.shape)
print ('Shape of testX :', testX.shape)
```

```python
#Create and fit the LSTM Neural Network:

model= Sequential() #sequential means in order. from layer 1 to 2...
model.add(LSTM(4, input_shape=(1, look_back)))
model.add(Dense(1))

#in keras.io website, under API/layers, you will find all the layers and info

model.compile (loss= 'mean_squared_error', optimizer='adam')
model.fit(trainX, trainY, epochs=500, batch_size=32, verbose=2) #going 500 times through all data so takes time to give output
```

```python
# Making predictions:

trainPredict =model.predict(trainX) #doing this to compare the error we have on the values we programmed the model with to the new values the model has never seen before.  
testPredict = model.predict(testX)

#Invert predictions: #to get rid of some of the bias so that we can level the field more

trainPredict = np.expm1(trainPredict)
trainY = np.expm1(trainY) #expm1=exponential m1
testPredict= np.expm1(testPredict)
testY= np.expm1(testY)
#reformatting data so that it all matches

#Calculate the root mean squared error:

trainScore = math.sqrt(mean_squared_error(trainY, trainPredict[:,0]))
print ('Train Score: %.2f RMSE' % (trainScore)) #RMSE = root mean square error
testScore = math.sqrt(mean_squared_error(testY, testPredict[:,0]))
print ('Test Score: %.2f RMSE' % (testScore))

#In our output : train score = 2.42 of the root mean sqaure error , test score = 3.17 of root mean square error. that means test score has more error. 
#IF TEST SCORE> TRAIN SCORE THEN WE HAVE OVERTRAINED. SHOULD NEVER BE THAT WAY
```

```python
test_series = pd.Series(testY)

#State of model performance:

if testScore < test_series.std():
  print('\n[Model performance is GOOD ENOUGH!]')
  print('\nRMSE of test prediction < Standard Deviation of test dataset')
  print('%.2f' % (testScore), '<', '%.2f' % (test_series.std()))
else:
  print('\n[Model performance is NOT GOOD ENOUGH!]')
  print('\nRMSE of test prediction > Standard Deviation of test dataset')
  print('%.2f' % (testScore), '>', '%.2f' % (test_series.std()))
```
